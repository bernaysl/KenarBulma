Kenar algilama, nesne tespitinde kullanilan en temel yontemdir.
Piksel degerlerinde keskin bir gecis var ise burada kenar var demektir. Esikleme yontemini kullanarak hangilerinin gercekten kenar olduguna karar verecegiz.

Canny, kenar cikartma icin kullanilan bir algoritmadir.
cv.Canny(goruntu, min_esik,max_esik)
max esit'in ustunde kalanlar kesin kenar olacak. min altinda kalanlar kenar olarak degerlendirilmeyecek.

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread("/content/sudoku.jpg",0) # imread fonksiyonuna 0 parametresini verirsek direkt olarak siyah beyaz olarak okur.
#kenar bulmada renklerle isimiz olmadigi icin siyah beyaz goruntu kullaniyoruz.

kenar= cv.Canny(img,50,90)
```

```python
plt.figure(figsize=(12,14))
plt.subplot(121), plt.imshow(img), plt.title("orijinal")
plt.xticks([]), plt.yticks([])

plt.subplot(122), plt.imshow(kenar), plt.title("kenarlar")
plt.xticks([]), plt.yticks([])

plt.show()
```

```python
yuz_img= cv.imread("/content/kız cocugu.png",0)
yuz_kenar_img= cv.Canny(yuz_img,50,100)
```

```python
plt.figure(figsize=(12,16))
plt.subplot(121), plt.imshow(yuz_img, cmap="gray"), plt.title("orijinal")
plt.xticks([]), plt.yticks([])

plt.subplot(122), plt.imshow(yuz_kenar_img, cmap="gray"), plt.title("kenar")
plt.xticks([]), plt.yticks([])

plt.show()
```
